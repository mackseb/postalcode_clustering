```python
import pandas as pd
import numpy as np

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

# Postal Codes of Canada Clustering

```python
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
df = pd.read_html(link)[0]
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Borough"] != "Not assigned"]
```

```python
df
```

```python
df = df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
```

```python
df
```

```python
df.loc[df["Neighbourhood"] == "Not assigned","Neighbourhood"] = df[df["Neighbourhood"] == "Not assigned"]["Borough"]
```

```python
df
```

```python
df.shape
```

```python
geo = pd.read_csv("Geospatial_Coordinates.csv")
```

```python
geo.rename(columns={"Postal Code": "Postcode"}, inplace=True)
```

```python
geo
```

```python
df = df.merge(geo, on="Postcode")
```

```python
df
```

```python
import folium
from sklearn.cluster import KMeans
from geopy.geocoders import Nominatim
```

```python
# set latitude, longitude values
latitude, longitude = 43.727929, -79.262029
```

```python
# create map of New York using latitude and longitude values
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```

```python
X = df[["Latitude", "Longitude"]]
```

```python
# set number of clusters
kclusters = 5

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(X)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
df_merged = df.merge(pd.Series(kmeans.labels_, name="Cluster"), left_index=True, right_index=True)
```

```python
df_merged
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_merged['Latitude'], df_merged['Longitude'], df_merged['Neighbourhood'], df_merged['Cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
